```python
import numpy as np

def compute_latent_semantics(data, k):
    # Extract even numbered images and their feature vectors
    even_data = data

    # Group data by label
    grouped_data = even_data.groupby('Image Label')

    # Dictionary to store latent semantics for each label
    latent_semantics_dict = {}

    for label, group in grouped_data:
        # Extract feature vectors for the current label
      #  print('Label: ',label,'\ngroup: ', group)
        feature_vectors = np.array(group['Feature Vector Resnet50 AvgPool'].tolist())

        # Center the data (subtract mean)
        mean_vector = np.mean(feature_vectors, axis=0)
        centered_data = feature_vectors - mean_vector

        # Calculate the covariance matrix
        covariance_matrix = np.cov(centered_data, rowvar=False)

        # Compute eigenvalues and eigenvectors
        eigenvalues, eigenvectors = np.linalg.eigh(covariance_matrix)

        # Sort eigenvalues and corresponding eigenvectors in descending order
        idx = np.argsort(eigenvalues)[::-1]
        eigenvalues = eigenvalues[idx]
        eigenvectors = eigenvectors[:, idx]

        # Take the top k eigenvectors to form the latent semantics
        latent_semantics = eigenvectors[:, :k]

        # Store the latent semantics for the current label
        latent_semantics_dict[label] = latent_semantics

        # Print k latent semantics for each label
        print(f'Label: {label}, Latent Semantics:')
        print(latent_semantics)
        print('---')

    return latent_semantics_dict


def predict_labels(query_data, latent_semantics_dict):
    # Dictionary to store predictions for each label
    predictions_dict = {'True Label': [], 'Predicted Label': []}

    for index, row in query_data.iterrows():
        # Extract relevant information
        true_label = row['Image Label']
        feature_vector = np.array(row['Feature Vector Resnet50 AvgPool'])

        # Find the label-specific latent semantics
        latent_semantics = latent_semantics_dict.get(true_label, None)

        if latent_semantics is not None:
            # Project the feature vector onto the label-specific latent semantics
            projection = np.dot(latent_semantics.T, feature_vector)

            # Find the label with the closest projection
            predicted_label = min(latent_semantics_dict.keys(),
                                  key=lambda label: np.linalg.norm(projection - np.dot(latent_semantics_dict[label].T, feature_vector)))
            print('predicted label for image',index, ':' ,predicted_label)

            # Store predictions
            predictions_dict['True Label'].append(true_label)
            predictions_dict['Predicted Label'].append(predicted_label)

    return predictions_dict



def evaluate(predictions_dict):
    # Convert predictions to a DataFrame
    predictions_df = pd.DataFrame(predictions_dict)

    # Calculate overall accuracy
    overall_accuracy = (predictions_df['True Label'] == predictions_df['Predicted Label']).mean()

    # Calculate per-label precision, recall, and F1-score
    per_label_metrics = {}
    unique_labels = predictions_df['True Label'].unique()

    for label in unique_labels:
        true_positives = ((predictions_df['True Label'] == label) & (predictions_df['Predicted Label'] == label)).sum()
        false_positives = ((predictions_df['True Label'] != label) & (predictions_df['Predicted Label'] == label)).sum()
        false_negatives = ((predictions_df['True Label'] == label) & (predictions_df['Predicted Label'] != label)).sum()

        precision = true_positives / (true_positives + false_positives)
        recall = true_positives / (true_positives + false_negatives)
        f1_score = 2 * (precision * recall) / (precision + recall)

        per_label_metrics[label] = {'Precision': precision, 'Recall': recall, 'F1-Score': f1_score}

    return overall_accuracy, per_label_metrics
```

```python
import numpy as np
from itertools import combinations

def generate_lsh(num_layers, num_hashes, dimensions):
    tables = [{} for _ in range(num_layers)]

    # Generate random hash functions for each layer
    hash_functions = np.random.randn(num_layers, num_hashes, dimensions)

    def hash_vector(vector):
        hash_values = np.dot(hash_functions, vector)
        return (hash_values >= 0).astype(int)

    def index_vector(vector, index):
        hash_values = hash_vector(vector)
        for i, h in enumerate(hash_values):
            hash_key = tuple(h)
            if hash_key not in tables[i]:
                tables[i][hash_key] = []
            tables[i][hash_key].append(index)

    def build_index(vectors):
        for i, vector in enumerate(vectors):
            index_vector(vector, i)

    def query(query_vector, threshold=0.1):
        query_hash = hash_vector(query_vector)
        candidate_indices = set()

        for i, h in enumerate(query_hash):
            hash_key = tuple(h)
            if hash_key in tables[i]:
                candidate_indices.update(tables[i][hash_key])

        # Compute actual distances and filter by threshold
        candidates = [(idx, np.linalg.norm(query_vector - vectors[idx]))
                      for idx in candidate_indices]

        return sorted(candidates, key=lambda x: x[1])[:10]  # Return top 10

    return hash_vector, index_vector, build_index, query

# Example Usage:
# Create LSH index with 2 layers, 4 hashes per layer, and vectors of dimension 256
num_layers = 2
num_hashes = 4
dimensions = 256
hash_vector, index_vector, build_index, query = generate_lsh(num_layers, num_hashes, dimensions)

# Generate random vectors for testing
num_vectors = 1000
vector_dimensions = 256
vectors = np.random.rand(num_vectors, vector_dimensions)
print("vectors first \n len \n", vectors[0],"\n",len(vectors))

# Build the LSH index
build_index(vectors)

# Query a random vector and get similar vectors
query_vector = np.random.rand(vector_dimensions)
similar_vectors = query(query_vector)

print("Similar Vectors:")
for idx, distance in similar_vectors:
    print(f"Index: {idx}, Distance: {distance}")
```
